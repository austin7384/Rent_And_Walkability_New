```python
import pandas as pd
import numpy as np
import geopandas as gpd
from shapely.geometry import Point
```

```python
dfCrime = pd.read_csv('/Users/austincoffelt/Documents/Rent-Walk-Local/CrimeWatch_Data_20250422.csv')
print(dfCrime.shape)
dfCrime.head()
```

```python
dfCrime = dfCrime[(dfCrime['Location'].isna() == False)]
print(dfCrime.shape)
dfCrime.head()
```

```python
dfCrime['DateTime'] = pd.to_datetime(dfCrime['DateTime'])
dfCrime = dfCrime.loc[(dfCrime['DateTime'] >= '2019-01-01') & (dfCrime['DateTime'] < '2020-01-01')]
print(dfCrime.shape)
dfCrime.head()
```

```python
list(dfCrime['CrimeType'].unique())
```

```python
violentCrimes = ['HOMICIDE',
 'ROBBERY',
 'FORCIBLE RAPE',
 'FELONY ASSAULT', 'ATTEMPTED RAPE', 'OTHER SEX OFFENSES']
dfCrime = dfCrime[dfCrime['CrimeType'].isin(violentCrimes)]
print(dfCrime.shape)
dfCrime.head()
```

```python
gdfPoints = gpd.GeoDataFrame(dfCrime, geometry=gpd.GeoSeries.from_wkt(dfCrime['Location']))
gdfPoints
```

```python
dfCrime = gdfPoints[['CrimeType', 'geometry']]
dfCrime.head()
```

```python
# Read shapefile (automatically loads .shp, .shx, .dbf)
gdf = gpd.read_file("/Users/austincoffelt/Documents/Rent-Walk-Local/cb_2018_06_bg_500k/cb_2018_06_bg_500k.shp")

# View the first 5 rows
print(gdf.shape)
gdf.head()
```

```python
# make sure points match the crs of gdf
gdfPoints = gpd.GeoDataFrame(
    dfCrime, 
    geometry='geometry',
    crs="EPSG:4269"  # WGS84 coordinate reference system
)
gdfPoints.head()
```

```python
# Basic spatial join (points within polygons)
result = gpd.sjoin(
    gdfPoints, 
    gdf,
    how="left",       # 'left', 'right', or 'inner'
    predicate="within" # Spatial relationship
)
print(result.shape)
result.head()
```

```python
crime = result['AFFGEOID'].value_counts()
crime
```

```python
# export as a csv
crime.to_csv('/Users/austincoffelt/Documents/RW Data/Oakland_Violent_Crime_BG.csv')
```
