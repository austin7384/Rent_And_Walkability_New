```python
import numpy as np
import pandas as pd
import geopandas as gpd
from shapely.geometry import Point
```

```python
dfCrime = pd.read_csv('/Users/tanisha/Desktop/Raleigh_Police_Incidents_(NIBRS).csv')
print(dfCrime.shape)
dfCrime.head()
```

```python
list(dfCrime.columns)
```

```python
dfCrime = dfCrime[dfCrime['reported_year'] == 2018]
print(dfCrime.shape)
dfCrime.head()
```

```python
list(dfCrime['crime_description'].unique())
```

```python
violentCrimes = [
    'Homicide/Murder-Nonnegligent Manslaughter',
    'Homicide/Negligent Manslaughter',
    'Homicide/Justifiable Homicide',
    'Robbery/From Person',
    'Robbery/From Business',
    'Assault/Aggravated',
    'Sex Offense/Forcible Rape',
    'Sex Offense/Forcible Fondling',
    'Sex Offense/Forcible Sodomy',
    'Sex Offense/Assault with an Object',
    'Child Abuse/Aggravated',
    'Kidnapping/Abduction',
    'Human Trafficking/Involuntary Servitude',
    'Human Trafficking/Commercial Sex Acts'
]

dfCrime = dfCrime[dfCrime['crime_description'].isin(violentCrimes)]
print(dfCrime.shape)
dfCrime.head()
```

```python
dfCrime = dfCrime[['GlobalID', 'latitude', 'longitude']]

# make our lon and lat into a point type for spacial join
dfCrime['geometry'] = dfCrime.apply(lambda row: Point(row['longitude'], row['latitude']), axis=1)
dfCrime = dfCrime[['GlobalID', 'geometry']]
dfCrime.head()
```

```python
# Read shapefile (automatically loads .shp, .shx, .dbf)
gdf = gpd.read_file("/Users/tanisha/Desktop/cb_2018_53_bg_500k/cb_2018_53_bg_500k.shp")

# View the first 5 rows
print(gdf.shape)
gdf.head()
```

```python
# make sure points match the crs of gdf
gdfPoints = gpd.GeoDataFrame(
    dfCrime, 
    geometry='geometry',
    crs="EPSG:4269"  # WGS84 coordinate reference system
)
gdfPoints.head()
```

```python
# Basic spatial join (points within polygons)
result = gpd.sjoin(
    gdfPoints, 
    gdf,
    how="left",       # 'left', 'right', or 'inner'
    predicate="within" # Spatial relationship
)
print(result.shape)
result.head()
```

```python
crime = result['AFFGEOID'].value_counts()
crime
```

```python
# export as a csv
crime.to_csv('seattle_Violent_Crime_BG.csv')
```
