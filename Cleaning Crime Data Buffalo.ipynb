```python
import pandas as pd
import numpy as np
import geopandas as gpd
from shapely.geometry import Point
```

```python
dfCrime = pd.read_csv('/Users/austincoffelt/Documents/Rent-Walk-Local/Crime_incidents_20250422.csv')
print(dfCrime.shape)
dfCrime.head()
```

```python
dfCrime['Incident Datetime'] = pd.to_datetime(dfCrime['Incident Datetime'])
dfCrime.head()
```

```python
dfCrime = dfCrime.loc[(dfCrime['Incident Datetime'] >= '2019-01-01') & (dfCrime['Incident Datetime'] < '2020-01-01')]
print(dfCrime.shape)
dfCrime.head()
```

```python
list(dfCrime.columns)
```

```python
sorted(list(dfCrime['Incident Type Primary'].unique()))
```

```python
violentCrimes = ['AGGR ASSAULT', 'RAPE', 'MURDER', 'SEXUAL ABUSE', 'ROBBERY']
dfCrime = dfCrime[dfCrime['Incident Type Primary'].isin(violentCrimes)]
print(dfCrime.shape)
dfCrime.head()
```

```python
dfCrime = dfCrime[['Incident Type Primary', 'Longitude', 'Latitude']]
dfCrime = dfCrime[dfCrime['Longitude'] != 'UNKNOWN']
dfCrime.head()
```

```python
# make our lon and lat into a point type for spacial join
dfCrime['geometry'] = dfCrime.apply(lambda row: Point(row['Longitude'], row['Latitude']), axis=1)
dfCrime = dfCrime[['Incident Type Primary', 'geometry']]
print(dfCrime.shape)
dfCrime.head()
```

```python
# Read shapefile (automatically loads .shp, .shx, .dbf)
gdf = gpd.read_file("/Users/austincoffelt/Documents/Rent-Walk-Local/cb_2018_36_bg_500k/cb_2018_36_bg_500k.shp")

# View the first 5 rows
print(gdf.shape)
gdf.head()
```

```python
# make sure points match the crs of gdf
gdfPoints = gpd.GeoDataFrame(
    dfCrime, 
    geometry='geometry',
    crs="EPSG:4269"  # WGS84 coordinate reference system
)
gdfPoints.head()
```

```python
# Basic spatial join (points within polygons)
result = gpd.sjoin(
    gdfPoints, 
    gdf,
    how="left",       # 'left', 'right', or 'inner'
    predicate="within" # Spatial relationship
)
print(result.shape)
result.head()
```

```python
crime = result['AFFGEOID'].value_counts()
crime
```

```python
# export as a csv
crime.to_csv('/Users/austincoffelt/Documents/RW Data/Buffalo_Violent_Crime_BG.csv')
```
