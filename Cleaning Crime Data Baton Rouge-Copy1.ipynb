```python
import pandas as pd
import numpy as np
import geopandas as gpd
from shapely.geometry import Point
```

```python
dfCrime = pd.read_csv('/Users/tanisha/Desktop/Baton_Rouge_Police_Crime_Incidents.csv')
print(dfCrime.shape)
dfCrime.head()
```

```python
dfCrime = dfCrime[(dfCrime['LATITUDE'].isna() == False) & (dfCrime['LONGITUDE'].isna() == False)]
print(dfCrime.shape)
dfCrime.head()
```

```python
dfCrime['REPORT DATE'] = pd.to_datetime(dfCrime['REPORT DATE'])
dfCrime = dfCrime[(dfCrime['REPORT DATE'] >= '2021-01-01') & (dfCrime['REPORT DATE'] < '2022-01-01')]
print(dfCrime.shape)
dfCrime.head()
```

```python
dfCrime.columns
```

```python
list(dfCrime['NIBRS CODE'].unique())
```

```python
violentCrimes = ['13A','09A','120']
dfCrime = dfCrime[dfCrime['NIBRS CODE'].isin(violentCrimes)]
print(dfCrime.shape)
dfCrime.head()
```

```python
dfCrime = dfCrime[['LATITUDE', 'LONGITUDE','NIBRS DESCRIPTION']]
dfCrime.head()
```

```python
# make our lon and lat into a point type for spacial join
dfCrime['geometry'] = dfCrime.apply(lambda row: Point(row['LONGITUDE'], row['LATITUDE']), axis=1)
dfCrime = dfCrime[['NIBRS DESCRIPTION', 'geometry']]
print(dfCrime.shape)
dfCrime.head()
```

```python
# Read shapefile (automatically loads .shp, .shx, .dbf)
gdf = gpd.read_file("/Users/tanisha/Downloads/cb_2022_us_bg_500k/cb_2022_us_bg_500k.shp")

# View the first 5 rows
print(gdf.shape)
gdf.head()
```

```python
# make sure points match the crs of gdf
gdfPoints = gpd.GeoDataFrame(
    dfCrime, 
    geometry='geometry',
    crs="EPSG:4269"  # WGS84 coordinate reference system
)
gdfPoints.head()
```

```python
# Basic spatial join (points within polygons)
result = gpd.sjoin(
    gdfPoints, 
    gdf,
    how="left",       # 'left', 'right', or 'inner'
    predicate="within" # Spatial relationship
)
print(result.shape)
result.head()
```

```python
crime = result['AFFGEOID'].value_counts()
crime
```

```python
# export as a csv
crime.to_csv('BR_Violent_Crime_BG.csv')
```
