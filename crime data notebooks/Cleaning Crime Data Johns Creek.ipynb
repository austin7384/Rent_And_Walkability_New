```python
import pandas as pd
import numpy as np
import geopandas as gpd
from shapely.geometry import Point
```

```python
dfCrime = pd.read_csv('/Users/austincoffelt/Documents/Rent-Walk-Local/Incident_20250422.csv')
print(dfCrime.shape)
dfCrime.head()
```

```python
list(dfCrime.columns)
```

```python
dfCrime['Date/Time Occurred'] = pd.to_datetime(dfCrime['Date/Time Occurred'])
dfCrime.head()
```

```python
dfCrime = dfCrime.loc[(dfCrime['Date/Time Occurred'] >= '2019-01-01') & (dfCrime['Date/Time Occurred'] < '2020-01-01')]
print(dfCrime.shape)
dfCrime.head()
```

```python
sorted(list(dfCrime['Offense'].unique()))
```

```python
violentCrimes = ['AGG ASSAULT/ AGG BATTERY-GUN', 'ROBBERY - OTHER', 'ROBBERY-BUSINESS-GUN', 'ROBBERY-BUSINESS-STRONGARM',
                'ROBBERY-FORCIBLE PURSE SNATCHING', 'ROBBERY-STREET-GUN', 'SEX OFFENSE - OTHER', 'SEXUAL ASSAULT - OTHER', 'PERSON DEAD']
propertyCrimes = ['BURGLARY - NONRESIDENCE - FORCED ENTRY', 'BURGLARY - NONRESIDENCE - NO FORCE ENTRY', 'BURGLARY - OTHER',
                  'BURGLARY - RESIDENCE - FORCED ENTRY', 'BURGLARY - RESIDENCE - NO FORCED ENTRY', 'LARCENY - FROM BUILDING',
                  'LARCENY - FROM MAIL', 'LARCENY - FROM RESIDENTIAL MAILBOX', 'LARCENY - FROM VEHICLE', 'LARCENY - FROM YARDS',
                  'LARCENY - OTHER OFFENSES', 'LARCENY - PARTS FROM VEHICLE', 'LARCENY - POCKET PICKING', 'LARCENY - SHOPLIFTING',
                  'THEFT AND USE AUTOMOBILE FOR OTHER CRIME', 'THEFT BY EXTORTION', 'THEFT BY RECEIVING STOLEN PROPERTY', 'THEFT BY TAKING MOTOR VEHICLE',
                  'THEFT OF OTHER VEHICLES']
both = violentCrimes + propertyCrimes
dfCrime = dfCrime[dfCrime['Offense'].isin(both)]
dfCrime = dfCrime[['Offense', 'geox', 'geoy']]
print(dfCrime.shape)
dfCrime.head()
```

```python
# make our lon and lat into a point type for spacial join
dfCrime['geometry'] = dfCrime.apply(lambda row: Point(row['geoy'], row['geox']), axis=1)
dfCrime = dfCrime[['Offense', 'geometry']]
print(dfCrime.shape)
dfCrime.head()
```

```python
# Read shapefile (automatically loads .shp, .shx, .dbf)
gdf = gpd.read_file("/Users/austincoffelt/Documents/Rent-Walk-Local/cb_2018_13_bg_500k/cb_2018_13_bg_500k.shp")

# View the first 5 rows
print(gdf.shape)
gdf.head()
```

```python
# make sure points match the crs of gdf
gdfPoints = gpd.GeoDataFrame(
    dfCrime, 
    geometry='geometry',
    crs="EPSG:4269"  # WGS84 coordinate reference system
)
gdfPoints.head()
```

```python
# Basic spatial join (points within polygons)
result = gpd.sjoin(
    gdfPoints, 
    gdf,
    how="left",       # 'left', 'right', or 'inner'
    predicate="within" # Spatial relationship
)
print(result.shape)
result.head()
```

```python
crime = result['AFFGEOID'].value_counts()
crime
```

```python
crime.to_csv('/Users/austincoffelt/Documents/RW Data/J_Creek_All_crime_by_BG.csv')
```
