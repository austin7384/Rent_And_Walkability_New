```python
import pandas as pd
import numpy as np
import geopandas as gpd
from shapely.geometry import Point
```

```python
dfCrime = pd.read_csv('/Users/austincoffelt/Downloads/Crime_Reports_20250418.csv')
print(dfCrime.shape)
dfCrime.head()
```

```python
dfCrime['Date of Report'] = pd.to_datetime(dfCrime['Date of Report'])
dfCrime.head()
```

```python
dfCrime = dfCrime.loc[(dfCrime['Date of Report'] >= '2018-01-01') & (dfCrime['Date of Report'] < '2019-01-01')]
print(dfCrime.shape)
dfCrime.head()
```

```python
sorted(list(dfCrime['Crime'].unique()))
```

```python
violentCrimes = ['Aggravated Assault', 'Commercial Robbery', 'Homicide', 'Street Robbery']
propertyCrimes = ['Auto Theft', 'Commercial Break', 'Housebreak', 'Larceny (Misc)', 'Larceny from Building',
                  'Larceny from MV', 'Larceny from Person', 'Larceny from Residence', 'Larceny of Bicycle',
                  'Larceny of Plate', 'Larceny of Services']
both = violentCrimes + propertyCrimes
dfCrime = dfCrime[dfCrime['Crime'].isin(both)]
print(dfCrime.shape)
dfCrime.head()
```

```python
import pandas as pd
from tqdm import tqdm
import requests
import time  # You need to import time for the sleep function

def geocode_address(address, api_key):
    base_url = "https://maps.googleapis.com/maps/api/geocode/json"
    params = {
        "address": address,
        "key": api_key
    }
    
    try:
        response = requests.get(base_url, params=params)
        response.raise_for_status()  # Raises an exception for HTTP errors
        data = response.json()
        
        if data["status"] == "OK":
            location = data["results"][0]["geometry"]["location"]
            return location["lat"], location["lng"]
        else:
            print(f"Geocoding failed for {address}. Status: {data['status']}")
            return None, None
    except Exception as e:
        print(f"Error geocoding {address}: {str(e)}")
        return None, None

def geocode_dataframe(df, api_key, address_col='Location'):
    results = []
    
    for address in tqdm(df[address_col], desc="Geocoding addresses"):
        lat, lng = geocode_address(address, api_key)
        results.append({
            'address': address,
            'latitude': lat,
            'longitude': lng,
            'geocoded': lat is not None
        })
        time.sleep(0.1)  # Respect rate limits
    
    return pd.DataFrame(results)

geocoded_df = geocode_dataframe(dfCrime, '1')
print(geocoded_df.shape)
geocoded_df.head()
```

```python
# make our lon and lat into a point type for spacial join
geocoded_df['geometry'] = geocoded_df.apply(lambda row: Point(row['longitude'], row['latitude']), axis=1)
dfCrime = geocoded_df[['address', 'geometry']]
dfCrime.head()
```

```python
# Read shapefile (automatically loads .shp, .shx, .dbf)
gdf = gpd.read_file("/Users/austincoffelt/Documents/Rent-Walk-Local/cb_2018_25_bg_500k/cb_2018_25_bg_500k.shp")

# View the first 5 rows
print(gdf.shape)
gdf.head()
```

```python
# make sure points match the crs of gdf
gdfPoints = gpd.GeoDataFrame(
    dfCrime, 
    geometry='geometry',
    crs="EPSG:4269"  # WGS84 coordinate reference system
)
gdfPoints.head()
```

```python
# Basic spatial join (points within polygons)
result = gpd.sjoin(
    gdfPoints, 
    gdf,
    how="left",       # 'left', 'right', or 'inner'
    predicate="within" # Spatial relationship
)
print(result.shape)
result.head()
```

```python
crime = result['AFFGEOID'].value_counts()
crime
```
