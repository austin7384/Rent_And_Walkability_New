```python
import pandas as pd
import numpy as np
import geopandas as gpd
from shapely.geometry import Point
```

```python
dfCrime = pd.read_csv('/Users/austincoffelt/Documents/Rent-Walk-Local/Crime_Data_from_2010_to_2019_20250419.csv')
print(dfCrime.shape)
dfCrime.head()
```

```python
sorted(list(dfCrime['Crm Cd'].unique()))
```

```python
dfCrime[dfCrime['Crm Cd'] == 520].head(1)
```

```python
violentCrimeCodes = [110, 113, 121, 122, 210, 220, 230, 231, 236]
propertyCrimes = [310, 320, 330, 331, 341, 343, 345, 347, 349, 350, 351, 352, 410, 470, 473, 480, 487, 510, 520]
both = violentCrimeCodes + propertyCrimes
dfCrime = dfCrime[dfCrime['Crm Cd'].isin(both)]
print(dfCrime.shape)
dfCrime.head()
```

```python
dfCrime = dfCrime[['Crm Cd', 'DATE OCC',  'LON', 'LAT']]
dfCrime.head()
```

```python
# make our lon and lat into a point type for spacial join
dfCrime['geometry'] = dfCrime.apply(lambda row: Point(row['LON'], row['LAT']), axis=1)
dfCrime = dfCrime[['Crm Cd', 'DATE OCC', 'geometry']]
print(dfCrime.shape)
dfCrime.head()
```

```python
dfCrime['DATE OCC'] = pd.to_datetime(dfCrime['DATE OCC'])
dfCrime18 = dfCrime[(dfCrime['DATE OCC'] >= '2019-01-01') & (dfCrime['DATE OCC'] < '2020-01-01')]
dfCrime17 = dfCrime[(dfCrime['DATE OCC'] >= '2017-01-01') & (dfCrime['DATE OCC'] < '2018-01-01')]
dfCrime16 = dfCrime[(dfCrime['DATE OCC'] >= '2016-01-01') & (dfCrime['DATE OCC'] < '2017-01-01')]
dfCrime15 = dfCrime[(dfCrime['DATE OCC'] >= '2015-01-01') & (dfCrime['DATE OCC'] < '2016-01-01')]
dfCrime14 = dfCrime[(dfCrime['DATE OCC'] >= '2014-01-01') & (dfCrime['DATE OCC'] < '2015-01-01')]
```

```python
# Read shapefile (automatically loads .shp, .shx, .dbf)
gdf = gpd.read_file("/Users/austincoffelt/Documents/Rent-Walk-Local/cb_2018_06_bg_500k/cb_2018_06_bg_500k.shp")

# View the first 5 rows
print(gdf.shape)
gdf.head()
```

```python
# make sure points match the crs of gdf
gdfPoints18 = gpd.GeoDataFrame(
    dfCrime18, 
    geometry='geometry',
    crs="EPSG:4269"  # WGS84 coordinate reference system
)
gdfPoints18.head()
```

```python
# Basic spatial join (points within polygons)
result18 = gpd.sjoin(
    gdfPoints18, 
    gdf,
    how="left",       # 'left', 'right', or 'inner'
    predicate="within" # Spatial relationship
)
print(result18.shape)
result18.head()
```

```python
crime18 = result18['AFFGEOID'].value_counts()
crime18
```

```python
# export as a csv
crime18.to_csv('/Users/austincoffelt/Documents/RW Data/LA_All_Crime_BG.csv')
```
