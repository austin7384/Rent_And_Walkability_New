```python
import pandas as pd
import numpy as np
import geopandas as gpd
from shapely.geometry import Point
```

```python
dfCrime = pd.read_csv('/Users/austincoffelt/Documents/Rent-Walk-Local/Police_Department_Incident_Reports__2018_to_Present_20250412.csv')
print(dfCrime.shape)
dfCrime.head()
```

```python
dfCrime = dfCrime[(dfCrime['Latitude'].isna() == False) & (dfCrime['Longitude'].isna() == False)]
print(dfCrime.shape)
dfCrime.head()
```

```python
dfCrime = dfCrime[dfCrime['Incident Year'] == 2019]
print(dfCrime.shape)
dfCrime.head()
```

```python
list(dfCrime['Incident Subcategory'].unique())
```

```python
violentCrimes = ['Aggravated Assault', 'Rape - Attempted', 'Robbery - Residential', 'Robbery - Street', 'Robbery - Carjacking',
                'Sex Offense', 'Robbery - Commercial', 'Rape', 'Homicide', 'Robbery - Other']
propertyCrimes = ['Motor Vehicle Theft', 'Larceny Theft - Other', 'Larceny - From Vehicle', 'Larceny Theft - Shoplifting',
                  'Larceny Theft - From Building', 'Theft From Vehicle', 'Burglary - Commercial', 'Burglary - Residential',
                  'Burglary - Other', 'Larceny Theft - Pickpocket', 'Burglary - Hot Prowl', 'Larceny Theft - Bicycle', 
                  'Larceny - Auto Parts', 'Larceny Theft - Purse Snatch', 'Motor Vehicle Theft (Attempted)']
both = violentCrimes + propertyCrimes
dfCrime = dfCrime[dfCrime['Incident Subcategory'].isin(both)]
print(dfCrime.shape)
dfCrime.head()
```

```python
dfCrime = dfCrime[['Incident Subcategory', 'Longitude', 'Latitude']]
dfCrime.head()
```

```python
# make our lon and lat into a point type for spacial join
dfCrime['geometry'] = dfCrime.apply(lambda row: Point(row['Longitude'], row['Latitude']), axis=1)
dfCrime = dfCrime[['Incident Subcategory', 'geometry']]
print(dfCrime.shape)
dfCrime.head()
```

```python
# Read shapefile (automatically loads .shp, .shx, .dbf)
gdf = gpd.read_file("/Users/austincoffelt/Documents/Rent-Walk-Local/cb_2018_06_bg_500k/cb_2018_06_bg_500k.shp")

# View the first 5 rows
print(gdf.shape)
gdf.head()
```

```python
# make sure points match the crs of gdf
gdfPoints = gpd.GeoDataFrame(
    dfCrime, 
    geometry='geometry',
    crs="EPSG:4269"  # WGS84 coordinate reference system
)
gdfPoints.head()
```

```python
# Basic spatial join (points within polygons)
result = gpd.sjoin(
    gdfPoints, 
    gdf,
    how="left",       # 'left', 'right', or 'inner'
    predicate="within" # Spatial relationship
)
print(result.shape)
result.head()
```

```python
crime = result['AFFGEOID'].value_counts()
crime
```

```python
# export as a csv
crime.to_csv('/Users/austincoffelt/Documents/RW Data/SF_All_Crime_BG.csv')
```
