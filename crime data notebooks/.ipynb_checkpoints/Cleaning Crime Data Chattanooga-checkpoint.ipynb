```python
import pandas as pd
import numpy as np
import geopandas as gpd
from shapely.geometry import Point
```

```python
dfCrime = pd.read_csv('/Users/austincoffelt/Documents/Rent-Walk-Local/Violent_Crime_Data_20250422.csv')
temp = pd.read_csv('/Users/austincoffelt/Documents/Rent-Walk-Local/All_Part_1_Property_Crimes_20250430.csv')
dfCrime = pd.concat([dfCrime, temp])
print(dfCrime.shape)
dfCrime.head()
```

```python
dfCrime['Date_Incident'] = pd.to_datetime(dfCrime['Date_Incident'])
dfCrime.head()
```

```python
dfCrime = dfCrime.loc[(dfCrime['Date_Incident'] >= '2019-01-01') & (dfCrime['Date_Incident'] < '2020-01-01')]
print(dfCrime.shape)
dfCrime.head()
```

```python
propertyAndViolentCrimes = ['13A', '09A', '120', '11A', '11B', '11C', '11D', '220',
                           '23A', '23B', '23C', '23D', '23E', '23F', '23G', '23H', '240']
dfCrime = dfCrime[dfCrime['UCR_Incident_Code'].isin(propertyAndViolentCrimes)]
print(dfCrime.shape)
dfCrime.head()
```

```python
dfCrime = dfCrime[['UCR_Incident_Code', 'Longitude', 'Latitude']]
dfCrime.head()
```

```python
# make our lon and lat into a point type for spacial join
dfCrime['geometry'] = dfCrime.apply(lambda row: Point(row['Longitude'], row['Latitude']), axis=1)
dfCrime = dfCrime[['UCR_Incident_Code', 'geometry']]
print(dfCrime.shape)
dfCrime.head()
```

```python
# Read shapefile (automatically loads .shp, .shx, .dbf)
gdf = gpd.read_file("/Users/austincoffelt/Documents/Rent-Walk-Local/cb_2018_47_bg_500k/cb_2018_47_bg_500k.shp")

# View the first 5 rows
print(gdf.shape)
gdf.head()
```

```python
# make sure points match the crs of gdf
gdfPoints = gpd.GeoDataFrame(
    dfCrime, 
    geometry='geometry',
    crs="EPSG:4269"  # WGS84 coordinate reference system
)
gdfPoints.head()
```

```python
# Basic spatial join (points within polygons)
result = gpd.sjoin(
    gdfPoints, 
    gdf,
    how="left",       # 'left', 'right', or 'inner'
    predicate="within" # Spatial relationship
)
print(result.shape)
result.head()
```

```python
crime = result['AFFGEOID'].value_counts()
crime
```

```python
# export as a csv
crime.to_csv('/Users/austincoffelt/Documents/RW Data/Chatta_All_Crime_BG.csv')
```
