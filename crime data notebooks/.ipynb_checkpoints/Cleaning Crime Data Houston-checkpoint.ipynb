```python
import pandas as pd
import numpy as np
import geopandas as gpd
from shapely.geometry import Point
```

```python
dfCrime = pd.read_excel('/Users/austincoffelt/Documents/Rent-Walk-Local/NIBRSPublicView2019.xlsx')
print(dfCrime.shape)
dfCrime.head()
```

```python
sorted(list(dfCrime['NIBRSDescription'].unique()))
```

```python
violentCrimes = ['Aggravated Assault', 'Forcible fondling', 'Forcible rape', 'Forcible sodomy', 'Murder, non-negligent',
                'Robbery']
propertyCrimes = ['All other larceny', 'Burglary, Breaking and Entering', 'Motor vehicle theft', 'Stolen property offenses',
                  'Theft from building', 'Theft from motor vehicle', 'Theft of motor vehicle parts or accessory']
both = violentCrimes + propertyCrimes
dfCrime = dfCrime[dfCrime['NIBRSDescription'].isin(both)]
print(dfCrime.shape)
dfCrime.head()
```

```python
dfCrime = dfCrime[['NIBRSDescription', 'MapLatitude', 'MapLongitude']]

# make our lon and lat into a point type for spacial join
dfCrime['geometry'] = dfCrime.apply(lambda row: Point(row['MapLongitude'], row['MapLatitude']), axis=1)
dfCrime = dfCrime[['NIBRSDescription', 'geometry']]
dfCrime.head()
```

```python
# Read shapefile (automatically loads .shp, .shx, .dbf)
gdf = gpd.read_file("/Users/austincoffelt/Documents/Rent-Walk-Local/cb_2018_48_bg_500k/cb_2018_48_bg_500k.shp")

# View the first 5 rows
print(gdf.shape)
gdf.head()
```

```python
# make sure points match the crs of gdf
gdfPoints = gpd.GeoDataFrame(
    dfCrime, 
    geometry='geometry',
    crs="EPSG:4269"  # WGS84 coordinate reference system
)
gdfPoints.head()
```

```python
# Basic spatial join (points within polygons)
result = gpd.sjoin(
    gdfPoints, 
    gdf,
    how="left",       # 'left', 'right', or 'inner'
    predicate="within" # Spatial relationship
)
print(result.shape)
result.head()
```

```python
crime = result['AFFGEOID'].value_counts()
crime
```

```python
# export as a csv
crime.to_csv('/Users/austincoffelt/Documents/RW Data/Houston_All_Crime_BG.csv')
```
