```python
import pandas as pd
import numpy as np
import geopandas as gpd
from shapely.geometry import Point
```

```python
dfCrime = pd.read_csv('/Users/austincoffelt/Documents/Rent-Walk-Local/tmp6w6ts2d7.csv')
print(dfCrime.shape)
dfCrime.head()
```

```python
sorted(list(dfCrime['OFFENSE_DESCRIPTION'].unique()))
```

```python
violentCrimes = ['ASSAULT - AGGRAVATED', 'ASSAULT - AGGRAVATED - BATTERY', 'MURDER, NON-NEGLIGIENT MANSLAUGHTER',
                'ROBBERY - BANK', 'ROBBERY - CAR JACKING', 'ROBBERY - COMMERCIAL', 'ROBBERY - HOME INVASION', 'ROBBERY - OTHER',
                'ROBBERY - STREET', 'ROBBERY - UNARMED - CHAIN STORE']
propertyCrimes = ['AUTO THEFT', 'AUTO THEFT - LEASED/RENTED VEHICLE', 'AUTO THEFT - MOTORCYCLE / SCOOTER', 'BURGLARY - COMMERICAL',
                  'BURGLARY - COMMERICAL - ATTEMPT', 'BURGLARY - COMMERICAL - FORCE', 'BURGLARY - COMMERICAL - NO FORCE',
                  'BURGLARY - OTHER - ATTEMPT', 'BURGLARY - OTHER - FORCE', 'BURGLARY - OTHER - NO FORCE', 'BURGLARY - RESIDENTIAL',
                  'BURGLARY - RESIDENTIAL - ATTEMPT', 'BURGLARY - RESIDENTIAL - FORCE', 'BURGLARY - RESIDENTIAL - NO FORCE', 'LARCENY ALL OTHERS',
                  'LARCENY PICK-POCKET', 'LARCENY PURSE SNATCH - NO FORCE', 'LARCENY PURSE SNATCH - NO FORCE ', 'LARCENY SHOPLIFTING',
                  'LARCENY THEFT FROM BUILDING', 'LARCENY THEFT FROM COIN-OP MACHINE', 'LARCENY THEFT FROM MV - NON-ACCESSORY',
                  'LARCENY THEFT OF BICYCLE', 'LARCENY THEFT OF MV PARTS & ACCESSORIES']
both = violentCrimes + propertyCrimes
dfCrime = dfCrime[dfCrime['OFFENSE_DESCRIPTION'].isin(both)]
print(dfCrime.shape)
dfCrime.head()
```

```python
dfCrime = dfCrime[['OFFENSE_DESCRIPTION', 'Lat', 'Long']]
dfCrime = dfCrime[dfCrime['Lat'].isna() == False]
dfCrime = dfCrime[dfCrime['Long'].isna() == False]

# make our lon and lat into a point type for spacial join
dfCrime['geometry'] = dfCrime.apply(lambda row: Point(row['Long'], row['Lat']), axis=1)
dfCrime = dfCrime[['OFFENSE_DESCRIPTION', 'geometry']]
dfCrime.head()
```

```python
# Read shapefile (automatically loads .shp, .shx, .dbf)
gdf = gpd.read_file("/Users/austincoffelt/Documents/Rent-Walk-Local/cb_2018_25_bg_500k/cb_2018_25_bg_500k.shp")

# View the first 5 rows
print(gdf.shape)
gdf.head()
```

```python
# make sure points match the crs of gdf
gdfPoints = gpd.GeoDataFrame(
    dfCrime, 
    geometry='geometry',
    crs="EPSG:4269"  # WGS84 coordinate reference system
)
gdfPoints.head()
```

```python
# Basic spatial join (points within polygons)
result = gpd.sjoin(
    gdfPoints, 
    gdf,
    how="left",       # 'left', 'right', or 'inner'
    predicate="within" # Spatial relationship
)
print(result.shape)
result.head()
```

```python
crime = result['AFFGEOID'].value_counts()
crime
```

```python
# export as a csv
crime.to_csv('/Users/austincoffelt/Documents/RW Data/Boston_All_Crime_BG.csv')
```

```python
result2 = result[result['OFFENSE_DESCRIPTION'].isin(violentCrimes)]
print(result2.shape)
result2.head()
```

```python
crime = result2['AFFGEOID'].value_counts()

# export as a csv
crime.to_csv('/Users/austincoffelt/Documents/RW Data/Boston_Violent_Crime_BG.csv')
```
