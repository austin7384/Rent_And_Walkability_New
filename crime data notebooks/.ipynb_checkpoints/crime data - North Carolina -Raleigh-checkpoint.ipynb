```python
import numpy as np
import pandas as pd
import geopandas as gpd
from shapely.geometry import Point
```

```python
dfCrime = pd.read_csv('/Users/austincoffelt/Documents/Rent-Walk-Local/Raleigh_Police_Incidents_(NIBRS).csv')
print(dfCrime.shape)
dfCrime.head()
```

```python
list(dfCrime.columns)
```

```python
dfCrime = dfCrime[dfCrime['reported_year'] == 2019]
print(dfCrime.shape)
dfCrime.head()
```

```python
list(dfCrime['crime_description'].unique())
```

```python
violentCrimes = [
    'Homicide/Murder-Nonnegligent Manslaughter',
    'Homicide/Negligent Manslaughter',
    'Robbery/From Person',
    'Robbery/From Business',
    'Assault/Aggravated',
    'Sex Offense/Forcible Rape',
    'Sex Offense/Forcible Fondling',
    'Sex Offense/Forcible Sodomy',
    'Sex Offense/Assault with an Object',
    'Child Abuse/Aggravated',
    'Human Trafficking/Involuntary Servitude']
propertyCrimes = ['Motor Vehicle Theft', 'Burglary/Residential', 'Burglary/Commercial or Non-Residential', 'Larceny/All Other',
                  'Larceny/Theft from Motor Vehicle', 'Larceny/Theft from Building', 'Larceny/Purse-Snatching',
                  'Larceny/Theft of MV Parts-Accessories', 'Larceny/Shoplifting', 'Larceny/Pocket-Picking']
both = violentCrimes + propertyCrimes
dfCrime = dfCrime[dfCrime['crime_description'].isin(both)]
print(dfCrime.shape)
dfCrime.head()
```

```python
dfCrime = dfCrime[['crime_description', 'latitude', 'longitude']]

# make our lon and lat into a point type for spacial join
dfCrime['geometry'] = dfCrime.apply(lambda row: Point(row['longitude'], row['latitude']), axis=1)
dfCrime = dfCrime[['crime_description', 'geometry']]
dfCrime.head()
```

```python
# Read shapefile (automatically loads .shp, .shx, .dbf)
gdf = gpd.read_file("/Users/austincoffelt/Documents/Rent-Walk-Local/cb_2018_37_bg_500k/cb_2018_37_bg_500k.shp")

# View the first 5 rows
print(gdf.shape)
gdf.head()
```

```python
# make sure points match the crs of gdf
gdfPoints = gpd.GeoDataFrame(
    dfCrime, 
    geometry='geometry',
    crs="EPSG:4269"  # WGS84 coordinate reference system
)
gdfPoints.head()
```

```python
# Basic spatial join (points within polygons)
result = gpd.sjoin(
    gdfPoints, 
    gdf,
    how="left",       # 'left', 'right', or 'inner'
    predicate="within" # Spatial relationship
)
print(result.shape)
result.head()
```

```python
crime = result['AFFGEOID'].value_counts()
crime
```

```python
# export as a csv
crime.to_csv('/Users/austincoffelt/Documents/RW Data/ral_All_Crime_BG.csv')
```
