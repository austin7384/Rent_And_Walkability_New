```python
import pandas as pd
import numpy as np
import geopandas as gpd
from shapely.geometry import Point
```

```python
dfCrime = pd.read_csv('/Users/austincoffelt/Documents/Rent-Walk-Local/Incidents_2019.csv')
print(dfCrime.shape)
dfCrime.head()
```

```python
list(dfCrime.columns)
```

```python
sorted(list(dfCrime['Statute/Ordinance Description'].unique()))
```

```python
violentCrimes = ['ASSAULT - FELONY - AGGRAVATED', 'ASSAULT - FELONY - AGGRAVATED -ATTEMPT', 'HOMICIDE', 
                 'HOMICIDE - ATTEMPT', 'ROBBERY       -ATTEMPT', 'ROBBERY -FIREARM', 'ROBBERY -FIREARM              -ATTEMPT',
                 'ROBBERY -KNIFE/CUTTING INSTR  -ATTEMPT', 'ROBBERY -KNIFE/CUTTING INSTRUMENT', 'ROBBERY -OTHER DANGEROUS WEAPON',
                 'ROBBERY -OTHER WEAPON         -ATTEMPT', 'ROBBERY -STRONG ARM -NOWEAPON', 'SEXUAL ASSAULT - CONTACT',
                 'SEXUAL ASSAULT-ATT PENETRATION', 'SEXUAL ASSAULT-PENETRATION']
propertyCrimes = ['BURGLARY      -ATTEMPT', 'BURGLARY-FORCIBLE ENTRY', 'BURGLARY-NO FORCE ENTRY', 'THEFT BY DECEPTION      $500/LESS',
                  'THEFT BY DECEPTION     $1500-$4999', 'THEFT BY DECEPTION     $5000/MORE', 'THEFT BY DECEPTION     $501 -$1499',
                  'THEFT BY DECEPTION   $500/LESS         EMBEZ', 'THEFT BY DECEPTION  $1500-$4999        EMBEZ', 'THEFT BY DECEPTION  $500/LESS -ATTEMPT',
                  'THEFT BY DECEPTION  $5000/MORE         EMBEZ', 'THEFT BY DECEPTION  $501-$1499         EMBEZ',
                  'THEFT BY DECEPTION $1500-4999 -ATTEMPT', 'THEFT BY DECEPTION $501-$1499 -ATTEMPT', 'THEFT BY EXTORTION          $500/LESS',
                  'THEFT BY EXTORTION         $1500-$4999', 'THEFT BY EXTORTION         $501-$1499', 'THEFT BY EXTORTION   $500/LESS  -ATMPT',
                  'THEFT BY EXTORTION  $5000/MORE  -ATMPT', 'THEFT BY EXTORTION  $501-$1499  -ATMPT', 'THEFT BY SHOPLIFT  $1500-4999',
                  'THEFT BY SHOPLIFT  $1500-4999 -ATTEMPT', 'THEFT BY SHOPLIFT  $500/LESS', 'THEFT BY SHOPLIFT  $500/LESS  -ATTEMPT',
                  'THEFT BY SHOPLIFT  $5000/MORE', 'THEFT BY SHOPLIFT  $501-$1499', 'THEFT BY SHOPLIFT  $501-$1499 -ATTEMPT',
                  'THEFT BY UNLAWFUL TAKING    $200/LESS', 'THEFT BY UNLAWFUL TAKING    $201-$499', 'THEFT BY UNLAWFUL TAKING    $50 - $199',
                  'THEFT BY UNLAWFUL TAKING    $500-1500', 'THEFT BY UNLAWFUL TAKING    - ATTEMPT', 'THEFT BY UNLAWFUL TAKING    OVR $1500',
                  'THEFT BY UNLAWFUL TAKING    UNDER $50', 'THEFT BY UNLAWFUL TAKING   $200-$1499', 'THEFT OF MOTOR VEHICLE        -ATTEMPT  MV',
                  'THEFT OF MOTOR VEHICLE - FELONY         MV', 'THEFT OF MOTOR VEHICLE - MISDEMEANOR    MV',
                  'THEFT OF SERVICES           $500/LESS', 'THEFT OF SERVICES          $501-$1499', 'THEFT OF SERVICES     $1500-$4999', 'THEFT OF TELECOMMUNICATIONS SERVICES', 'THEFT OF UTILITY SERVICES']
both = violentCrimes + propertyCrimes
dfCrime = dfCrime[dfCrime['Statute/Ordinance Description'].isin(both)]
dfCrime = dfCrime[['Statute/Ordinance Description', 'Occurred Block LON', 'Occurred Block LAT']]
print(dfCrime.shape)
dfCrime.head()
```

```python
# make our lon and lat into a point type for spacial join
dfCrime['geometry'] = dfCrime.apply(lambda row: Point(row['Occurred Block LON'], row['Occurred Block LAT']), axis=1)
dfCrime = dfCrime[['Statute/Ordinance Description', 'geometry']]
print(dfCrime.shape)
dfCrime.head()
```

```python
# Read shapefile (automatically loads .shp, .shx, .dbf)
gdf = gpd.read_file("/Users/austincoffelt/Documents/Rent-Walk-Local/cb_2018_31_bg_500k/cb_2018_31_bg_500k.shp")

# View the first 5 rows
print(gdf.shape)
gdf.head()
```

```python
# make sure points match the crs of gdf
gdfPoints = gpd.GeoDataFrame(
    dfCrime, 
    geometry='geometry',
    crs="EPSG:4269"  # WGS84 coordinate reference system
)
gdfPoints.head()
```

```python
# Basic spatial join (points within polygons)
result = gpd.sjoin(
    gdfPoints, 
    gdf,
    how="left",       # 'left', 'right', or 'inner'
    predicate="within" # Spatial relationship
)
print(result.shape)
result.head()
```

```python
crime = result['AFFGEOID'].value_counts()
crime
```

```python
crime.to_csv('/Users/austincoffelt/Documents/RW Data/Omaha_all_crime_by_BG.csv')
```
