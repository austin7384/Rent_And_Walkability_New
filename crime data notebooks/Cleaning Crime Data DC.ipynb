```python
import pandas as pd
import numpy as np
import geopandas as gpd
from shapely.geometry import Point
```

```python
dfCrime = pd.read_csv('/Users/austincoffelt/Documents/Rent-Walk-Local/Crime_Incidents_in_2019.csv')
print(dfCrime.shape)
dfCrime.head()
```

```python
list(dfCrime.columns)
```

```python
list(dfCrime['OFFENSE'].unique())
```

```python
violentCrimes = ['ASSAULT W/DANGEROUS WEAPON', 'SEX ABUSE', 'ROBBERY', 'HOMICIDE']
propertyCrimes = ['THEFT/OTHER', 'THEFT F/AUTO', 'ROBBERY', 'MOTOR VEHICLE THEFT', 'BURGLARY']
both = violentCrimes + propertyCrimes
dfCrime = dfCrime[dfCrime['OFFENSE'].isin(both)]
print(dfCrime.shape)
dfCrime.head()
```

```python
# make our lon and lat into a point type for spacial join
dfCrime['geometry'] = dfCrime.apply(lambda row: Point(row['LONGITUDE'], row['LATITUDE']), axis=1)
dfCrime = dfCrime[['OFFENSE', 'geometry']]
print(dfCrime.shape)
dfCrime.head()
```

```python
# Read shapefile (automatically loads .shp, .shx, .dbf)
gdf = gpd.read_file("/Users/austincoffelt/Documents/Rent-Walk-Local/cb_2018_11_bg_500k/cb_2018_11_bg_500k.shp")

# View the first 5 rows
print(gdf.shape)
gdf.head()
```

```python
# make sure points match the crs of gdf
gdfPoints = gpd.GeoDataFrame(
    dfCrime, 
    geometry='geometry',
    crs="EPSG:4269"  # WGS84 coordinate reference system
)
gdfPoints.head()
```

```python
# Basic spatial join (points within polygons)
result = gpd.sjoin(
    gdfPoints, 
    gdf,
    how="left",       # 'left', 'right', or 'inner'
    predicate="within" # Spatial relationship
)
print(result.shape)
result.head()
```

```python
crime = result['AFFGEOID'].value_counts()
crime
```

```python
# export as a csv
crime.to_csv('/Users/austincoffelt/Documents/RW Data/DC_All_Crime_BG.csv')
```
