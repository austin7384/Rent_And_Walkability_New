```python
import pandas as pd
import numpy as np
import geopandas as gpd
from shapely.geometry import Point
```

```python
dfCrime = pd.read_csv('/Users/austincoffelt/Documents/Rent-Walk-Local/PDI__Police_Data_Initiative__Crime_Incidents_20250422.csv')
print(dfCrime.shape)
dfCrime.head()
```

```python
dfCrime['DATE_REPORTED'] = pd.to_datetime(dfCrime['DATE_REPORTED'])
dfCrime = dfCrime[(dfCrime['DATE_REPORTED'] >= '2019-01-01') & (dfCrime['DATE_REPORTED'] < '2020-01-01')]
print(dfCrime.shape)
dfCrime.head()
```

```python
list(dfCrime.columns)
```

```python
list(dfCrime['THEFT_CODE'].unique())
```

```python
list(dfCrime['OFFENSE'].unique())
```

```python
violentCrimes = ['RAPE', 'ROBBERY', 'AGGRAVATED ROBBERY', 'MURDER', 'AGGRAVATED ASSAULT', 'AGGRAVATED VEHICULAR ASSAULT', 'AGGRAVATED MURDER', 'AGG VEH HOM/VEH HOM/VEH MNSLGHTER']
propertyCrimes = ['THEFT', 'BURGLARY', 'BREAKING AND ENTERING', 'AGGRAVATED ROBBERY']
both = violentCrimes + propertyCrimes
dfCrime = dfCrime[dfCrime['OFFENSE'].isin(both)]
print(dfCrime.shape)
dfCrime.head()
```

```python
dfCrime = dfCrime[['OFFENSE', 'THEFT_CODE', 'LONGITUDE_X', 'LATITUDE_X']]
dfCrime = dfCrime[dfCrime['LONGITUDE_X'].isna() == False]
dfCrime = dfCrime[dfCrime['LATITUDE_X'].isna() == False]
dfCrime.head()
```

```python
# make our lon and lat into a point type for spacial join
dfCrime['geometry'] = dfCrime.apply(lambda row: Point(row['LONGITUDE_X'], row['LATITUDE_X']), axis=1)
dfCrime = dfCrime[['OFFENSE', 'THEFT_CODE', 'geometry']]
print(dfCrime.shape)
dfCrime.head()
```

```python
# Read shapefile (automatically loads .shp, .shx, .dbf)
gdf = gpd.read_file("/Users/austincoffelt/Documents/Rent-Walk-Local/cb_2018_39_bg_500k/cb_2018_39_bg_500k.shp")

# View the first 5 rows
print(gdf.shape)
gdf.head()
```

```python
# make sure points match the crs of gdf
gdfPoints = gpd.GeoDataFrame(
    dfCrime, 
    geometry='geometry',
    crs="EPSG:4269"  # WGS84 coordinate reference system
)
gdfPoints.head()
```

```python
# Basic spatial join (points within polygons)
result = gpd.sjoin(
    gdfPoints, 
    gdf,
    how="left",       # 'left', 'right', or 'inner'
    predicate="within" # Spatial relationship
)
print(result.shape)
result.head()
```

```python
crime = result['AFFGEOID'].value_counts()
crime
```

```python
# export as a csv
crime.to_csv('/Users/austincoffelt/Documents/RW Data/Cinci_All_Crime_BG.csv')
```
