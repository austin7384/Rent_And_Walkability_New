```python
import pandas as pd
import numpy as np
import geopandas as gpd
from shapely.geometry import Point
```

```python
dfCrime = pd.read_csv('/Users/austincoffelt/Documents/Rent-Walk-Local/2019CrimePhilly.csv')
print(dfCrime.shape)
dfCrime.head()
```

```python
dfCrime['dispatch_date'] = pd.to_datetime(dfCrime['dispatch_date'])
dfCrime = dfCrime[(dfCrime['dispatch_date'] >= '2019-01-01') & (dfCrime['dispatch_date'] < '2020-01-01')]
print(dfCrime.shape)
dfCrime.head()
```

```python
list(dfCrime['text_general_code'].unique())
```

```python
violentCrimes = ['Robbery No Firearm', 'Aggravated Assault No Firearm', 'Robbery Firearm', 'Aggravated Assault Firearm',
                 'Rape', 'Homicide - Criminal']
dfCrime = dfCrime[dfCrime['text_general_code'].isin(violentCrimes)]
print(dfCrime.shape)
dfCrime.head()
```

```python
dfCrime = dfCrime[['text_general_code',  'lng', 'lat']]
dfCrime.head()
```

```python
# make our lon and lat into a point type for spacial join
dfCrime['geometry'] = dfCrime.apply(lambda row: Point(row['lng'], row['lat']), axis=1)
dfCrime = dfCrime[['text_general_code', 'geometry']]
print(dfCrime.shape)
dfCrime.head()
```

```python
# Read shapefile (automatically loads .shp, .shx, .dbf)
gdf = gpd.read_file("/Users/austincoffelt/Documents/Rent-Walk-Local/cb_2018_42_bg_500k/cb_2018_42_bg_500k.shp")

# View the first 5 rows
print(gdf.shape)
gdf.head()
```

```python
# make sure points match the crs of gdf
gdfPoints = gpd.GeoDataFrame(
    dfCrime, 
    geometry='geometry',
    crs="EPSG:4269"  # WGS84 coordinate reference system
)
gdfPoints.head()
```

```python
# Basic spatial join (points within polygons)
result = gpd.sjoin(
    gdfPoints, 
    gdf,
    how="left",       # 'left', 'right', or 'inner'
    predicate="within" # Spatial relationship
)
print(result.shape)
result.head()
```

```python
crime = result['AFFGEOID'].value_counts()
crime
```

```python
# export as a csv
crime.to_csv('/Users/austincoffelt/Documents/RW Data/Philly_Violent_Crime_BG.csv')
```
