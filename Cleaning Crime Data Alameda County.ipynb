```python
import pandas as pd
import numpy as np
import geopandas as gpd
from shapely.geometry import Point
```

```python
dfCrime = pd.read_csv('/Users/austincoffelt/Documents/Rent-Walk-Local/Crime_Reports_2012_Jul2022.csv')
print(dfCrime.shape)
dfCrime.head()
```

```python
dfCrime = dfCrime[(dfCrime['Latitude'].isna() == False) & (dfCrime['Longitude'].isna() == False)]
print(dfCrime.shape)
dfCrime.head()
```

```python
dfCrime['DateTime'] = pd.to_datetime(dfCrime['DateTime'])
dfCrime = dfCrime.loc[(dfCrime['DateTime'] >= '2019-01-01') & (dfCrime['DateTime'] < '2020-01-01')]
print(dfCrime.shape)
dfCrime.head()
```

```python
list(dfCrime['CrimeCode'].unique())
```

```python
dfCrime[dfCrime['CrimeCode'] == '13A']
```

```python
violentCrimes = ['13A', '09A', '120', '11A', '11B', '11C', '11D']
dfCrime = dfCrime[dfCrime['CrimeCode'].isin(violentCrimes)]
print(dfCrime.shape)
dfCrime.head()
```

```python
dfCrime = dfCrime[['CrimeCode', 'Longitude', 'Latitude']]
dfCrime.head()
```

```python
# make our lon and lat into a point type for spacial join
dfCrime['geometry'] = dfCrime.apply(lambda row: Point(row['Longitude'], row['Latitude']), axis=1)
dfCrime = dfCrime[['CrimeCode', 'geometry']]
print(dfCrime.shape)
dfCrime.head()
```

```python
# Read shapefile (automatically loads .shp, .shx, .dbf)
gdf = gpd.read_file("/Users/austincoffelt/Documents/Rent-Walk-Local/cb_2018_06_bg_500k/cb_2018_06_bg_500k.shp")

# View the first 5 rows
print(gdf.shape)
gdf.head()
```

```python
# make sure points match the crs of gdf
gdfPoints = gpd.GeoDataFrame(
    dfCrime, 
    geometry='geometry',
    crs="EPSG:4269"  # WGS84 coordinate reference system
)
gdfPoints.head()
```

```python
# Basic spatial join (points within polygons)
result = gpd.sjoin(
    gdfPoints, 
    gdf,
    how="left",       # 'left', 'right', or 'inner'
    predicate="within" # Spatial relationship
)
print(result.shape)
result.head()
```

```python
crime = result['AFFGEOID'].value_counts()
crime
```

```python
# export as a csv
crime.to_csv('/Users/austincoffelt/Documents/RW Data/Alameda_County_Violent_Crime_BG.csv')
```
