```python
import pandas as pd
import numpy as np
```

```python
dfCrime = pd.read_csv('/Users/austincoffelt/Downloads/Crime_Reports.csv')
print(dfCrime.shape)
dfCrime.head()
```

```python
dfCrime['Occurred Date'] = pd.to_datetime(dfCrime['Occurred Date'])
dfCrime.head()
```

```python
dfCrime = dfCrime.loc[(dfCrime['Occurred Date'] >= '2021-01-01') & (dfCrime['Occurred Date'] < '2022-01-01')]
print(dfCrime.shape)
dfCrime.head()
```

```python
list(dfCrime['Category Description'].unique())
```

```python
dfCrime = dfCrime[(dfCrime['Category Description'] == 'Aggravated Assault') |
                  (dfCrime['Category Description'] == 'Robbery') |
                  (dfCrime['Category Description'] == 'Rape') |
                  (dfCrime['Category Description'] == 'Murder')]
dfCrime.shape
```

```python
dfCrime = dfCrime[dfCrime['Census Block Group'].isnull() == False]
dfCrime['Census Block Group'] = dfCrime['Census Block Group'].apply(lambda x: f"{x:.0f}")
dfCrime['Census Block Group'] = '48' + dfCrime['Census Block Group']
dfCrime['AFFGEOID'] = '1500000US' + dfCrime['Census Block Group']
print(dfCrime.shape)
dfCrime.head()
```

```python
crime = dfCrime['AFFGEOID'].value_counts()
crime = pd.DataFrame(crime)
print(crime.shape)
crime.head()
```

```python
crime.to_csv('AustinTX_violent_crime_by_BG.csv')
```
