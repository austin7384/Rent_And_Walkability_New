```python
import pandas as pd
import numpy as np
import geopandas as gpd
from shapely.geometry import Point
```

```python
dfCrime = pd.read_csv('/Users/austincoffelt/Downloads/Rent-Walk-Local/Crime_Responses_20250422.csv')
print(dfCrime.shape)
dfCrime.head()
```

```python
dfCrime['Offense Date'] = pd.to_datetime(dfCrime['Offense Date'])
dfCrime = dfCrime[(dfCrime['Offense Date'] >= '2018-01-01') & (dfCrime['Offense Date'] < '2019-01-01')]
print(dfCrime.shape)
dfCrime.head()
```

```python
sorted(list(dfCrime['Incident Type'].unique()))
```

```python
violentCrimes = ['Assault (aggravated)', 'Dating Violence Aggravated Assault', 'Domestic Aggravated Assualt', 
                 'Homicide', 'Lewd or Lascivious Molestation Offender 18 or Older', 'Lewd/lasc. Act to a Child Under 16',
                'Robbery',
 'Robbery (armed)',
 'Robbery (carjacking)',
 'Robbery (home Invasion)',
 'Robbery (strong Arm)',
 'Robbery by Sudden Snatching',
                'Sexual Battery',
 'Sexual Battery With an Object']
dfCrime = dfCrime[dfCrime['Incident Type'].isin(violentCrimes)]
print(dfCrime.shape)
dfCrime.head()
```

```python
list(dfCrime.columns)
```

```python
dfCrime = dfCrime[['Address',  'Longitude', 'Latitude']]
dfCrime.head()
```

```python
# make our lon and lat into a point type for spacial join
dfCrime['geometry'] = dfCrime.apply(lambda row: Point(row['Longitude'], row['Latitude']), axis=1)
dfCrime = dfCrime[['Address', 'geometry']]
print(dfCrime.shape)
dfCrime.head()
```

```python
# Read shapefile (automatically loads .shp, .shx, .dbf)
gdf = gpd.read_file("/Users/austincoffelt/Downloads/Rent-Walk-Local/cb_2018_12_bg_500k/cb_2018_12_bg_500k.shp")

# View the first 5 rows
print(gdf.shape)
gdf.head()
```

```python
# make sure points match the crs of gdf
gdfPoints = gpd.GeoDataFrame(
    dfCrime, 
    geometry='geometry',
    crs="EPSG:4269"  # WGS84 coordinate reference system
)
gdfPoints.head()
```

```python
# Basic spatial join (points within polygons)
result = gpd.sjoin(
    gdfPoints, 
    gdf,
    how="left",       # 'left', 'right', or 'inner'
    predicate="within" # Spatial relationship
)
print(result.shape)
result.head()
```

```python
crime = result['AFFGEOID'].value_counts()
crime
```

```python
# export as a csv
crime.to_csv('Gaines_Violent_Crime_BG.csv')
```
