```python
import pandas as pd
import numpy as np
import geopandas as gpd
from shapely.geometry import Point
```

```python
dfCrime = pd.read_csv('/Users/austincoffelt/Downloads/Rent-Walk-Local/RMS_Crime_Incidents.csv')
print(dfCrime.shape)
dfCrime.head()
```

```python
list(dfCrime.columns)
```

```python
dfCrime = dfCrime[dfCrime['incident_year'] == 2018]
print(dfCrime.shape)
dfCrime.head()
```

```python
list(dfCrime['offense_category'].unique())
```

```python
violentCrimes = ['AGGRAVATED ASSAULT', 'SEXUAL ASSAULT', 'ROBBERY', 'HOMICIDE']
dfCrime = dfCrime[dfCrime['offense_category'].isin(violentCrimes)]
print(dfCrime.shape)
dfCrime.head()
```

```python
dfCrime = dfCrime[['offense_category', 'X', 'Y']]

# make our lon and lat into a point type for spacial join
dfCrime['geometry'] = dfCrime.apply(lambda row: Point(row['X'], row['Y']), axis=1)
dfCrime = dfCrime[['offense_category', 'geometry']]
dfCrime.head()
```

```python
# Read shapefile (automatically loads .shp, .shx, .dbf)
gdf = gpd.read_file("/Users/austincoffelt/Downloads/Rent-Walk-Local/cb_2018_26_bg_500k/cb_2018_26_bg_500k.shp")

# View the first 5 rows
print(gdf.shape)
gdf.head()
```

```python
# make sure points match the crs of gdf
gdfPoints = gpd.GeoDataFrame(
    dfCrime, 
    geometry='geometry',
    crs="EPSG:4269"  # WGS84 coordinate reference system
)
gdfPoints.head()
```

```python
# Basic spatial join (points within polygons)
result = gpd.sjoin(
    gdfPoints, 
    gdf,
    how="left",       # 'left', 'right', or 'inner'
    predicate="within" # Spatial relationship
)
print(result.shape)
result.head()
```

```python
crime = result['AFFGEOID'].value_counts()
crime
```

```python
# export as a csv
crime.to_csv('Detroit_Violent_Crime_BG.csv')
```
