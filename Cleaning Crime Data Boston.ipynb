```python
import pandas as pd
import numpy as np
import geopandas as gpd
from shapely.geometry import Point
```

```python
dfCrime = pd.read_csv('/Users/austincoffelt/Downloads/Rent-Walk-Local/tmpf_uzkqpk.csv')
print(dfCrime.shape)
dfCrime.head()
```

```python
violentCrimes = ['ASSAULT - AGGRAVATED', 'ASSAULT - AGGRAVATED - BATTERY', 'MURDER, NON-NEGLIGIENT MANSLAUGHTER',
                'ROBBERY - BANK', 'ROBBERY - CAR JACKING', 'ROBBERY - COMMERCIAL', 'ROBBERY - HOME INVASION', 'ROBBERY - OTHER',
                'ROBBERY - STREET', 'ROBBERY - UNARMED - CHAIN STORE', ]
dfCrime = dfCrime[dfCrime['OFFENSE_DESCRIPTION'].isin(violentCrimes)]
print(dfCrime.shape)
dfCrime.head()
```

```python
dfCrime = dfCrime[['OFFENSE_DESCRIPTION', 'Lat', 'Long']]

# make our lon and lat into a point type for spacial join
dfCrime['geometry'] = dfCrime.apply(lambda row: Point(row['Long'], row['Lat']), axis=1)
dfCrime = dfCrime[['OFFENSE_DESCRIPTION', 'geometry']]
dfCrime.head()
```

```python
# Read shapefile (automatically loads .shp, .shx, .dbf)
gdf = gpd.read_file("/Users/austincoffelt/Downloads/Rent-Walk-Local/cb_2018_25_bg_500k/cb_2018_25_bg_500k.shp")

# View the first 5 rows
print(gdf.shape)
gdf.head()
```

```python
# make sure points match the crs of gdf
gdfPoints = gpd.GeoDataFrame(
    dfCrime, 
    geometry='geometry',
    crs="EPSG:4269"  # WGS84 coordinate reference system
)
gdfPoints.head()
```

```python
# Basic spatial join (points within polygons)
result = gpd.sjoin(
    gdfPoints, 
    gdf,
    how="left",       # 'left', 'right', or 'inner'
    predicate="within" # Spatial relationship
)
print(result.shape)
result.head()
```

```python
crime = result['AFFGEOID'].value_counts()
crime
```

```python
# export as a csv
crime.to_csv('Boston_Violent_Crime_BG.csv')
```
