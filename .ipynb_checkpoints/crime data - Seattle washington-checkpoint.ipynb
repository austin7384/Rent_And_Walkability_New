```python
import numpy as np
import pandas as pd
import geopandas as gpd
from shapely.geometry import Point
```

```python
dfCrime = pd.read_csv('/Users/tanisha/Desktop/SPD_Crime_Data__2008-Present_20250422.csv')
print(dfCrime.shape)
dfCrime.head()
```

```python
list(dfCrime.columns)
```

```python
dfCrime['Report DateTime'] = pd.to_datetime(dfCrime['Report DateTime'])
dfCrime.head()
```

```python
dfCrime = dfCrime.loc[(dfCrime['Report DateTime'] >= '2018-01-01') & (dfCrime['Report DateTime'] < '2019-01-01')]
print(dfCrime.shape)
dfCrime.head()
```

```python
list(dfCrime['NIBRS Offense Code Description'].unique())
```

```python
violentCrimes = [
    'Murder & Nonnegligent Manslaughter',
    'Justifiable Homicide',
    'Rape',
    'Sexual Assault With An Object',
    'Fondling',
    'Sodomy',
    'Aggravated Assault',
    'Simple Assault',  # borderline but often tracked as violent
    'Robbery',
    'Kidnapping/Abduction',
    'Intimidation'
]

dfCrime = dfCrime[dfCrime['NIBRS Offense Code Description'].isin(violentCrimes)]
print(dfCrime.shape)
dfCrime.head()
```

```python
dfCrime = dfCrime[['Offense ID', 'Latitude', 'Longitude']]

# make our lon and lat into a point type for spacial join
dfCrime['geometry'] = dfCrime.apply(lambda row: Point(row['longitude'], row['latitude']), axis=1)
dfCrime = dfCrime[['GlobalID', 'geometry']]
dfCrime.head()
```

```python
# Read shapefile (automatically loads .shp, .shx, .dbf)
gdf = gpd.read_file("/Users/tanisha/Desktop/cb_2018_37_bg_500k/cb_2018_37_bg_500k.shp")

# View the first 5 rows
print(gdf.shape)
gdf.head()
```

```python
# make sure points match the crs of gdf
gdfPoints = gpd.GeoDataFrame(
    dfCrime, 
    geometry='geometry',
    crs="EPSG:4269"  # WGS84 coordinate reference system
)
gdfPoints.head()
```

```python
# Basic spatial join (points within polygons)
result = gpd.sjoin(
    gdfPoints, 
    gdf,
    how="left",       # 'left', 'right', or 'inner'
    predicate="within" # Spatial relationship
)
print(result.shape)
result.head()
```

```python
crime = result['AFFGEOID'].value_counts()
crime
```

```python
# export as a csv
crime.to_csv('Ral_Violent_Crime_BG.csv')
```
