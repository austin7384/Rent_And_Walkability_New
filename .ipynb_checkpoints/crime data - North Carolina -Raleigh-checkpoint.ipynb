```python
import numpy as np
import pandas as pd
import geopandas as gpd
from shapely.geometry import Point
```

```python
dfCrime = pd.read_csv('/Users/tanisha/Desktop/cpd-incidents.csv .csv')
print(dfCrime.shape)
dfCrime.head()
```

```python
list(dfCrime.columns)
```

```python
dfCrime['date_from'] = pd.to_datetime(dfCrime['date_from'])
dfCrime.head()
```

```python
dfCrime = dfCrime.loc[(dfCrime['date_from'] >= '2021-01-01') & (dfCrime['date_from'] < '2022-01-01')]
print(dfCrime.shape)
dfCrime.head()
```

```python
list(dfCrime['crime_type'].unique())
```

```python
violentCrimes = ['HOMICIDE - MANSLAUGHTER BY NEGLIGENCE', 'ROBBERY - ARMED', 'ROBBERY - COMMON LAW', 
                'ASSAULT - AGGRAVATED', 'SEX OFFENSES - ALL OTHER']

dfCrime = dfCrime[dfCrime['crime_type'].isin(violentCrimes)]
print(dfCrime.shape)
dfCrime.head()
```

```python
dfCrime = dfCrime[['ucr', 'lat', 'lon']]

# make our lon and lat into a point type for spacial join
dfCrime['geometry'] = dfCrime.apply(lambda row: Point(row['lon'], row['lat']), axis=1)
dfCrime = dfCrime[['ucr', 'geometry']]
dfCrime.head()
```

```python
# Read shapefile (automatically loads .shp, .shx, .dbf)
gdf = gpd.read_file("/Users/tanisha/Downloads/cb_2022_us_bg_500k/cb_2022_us_bg_500k.shp")

# View the first 5 rows
print(gdf.shape)
gdf.head()
```

```python
# make sure points match the crs of gdf
gdfPoints = gpd.GeoDataFrame(
    dfCrime, 
    geometry='geometry',
    crs="EPSG:4269"  # WGS84 coordinate reference system
)
gdfPoints.head()
```

```python
# Basic spatial join (points within polygons)
result = gpd.sjoin(
    gdfPoints, 
    gdf,
    how="left",       # 'left', 'right', or 'inner'
    predicate="within" # Spatial relationship
)
print(result.shape)
result.head()
```

```python
crime = result['AFFGEOID'].value_counts()
crime
```

```python
# export as a csv
crime.to_csv('Cary_Violent_Crime_BG.csv')
```
