```python
import numpy as np
import pandas as pd
import geopandas as gpd
from shapely.geometry import Point
```

```python
dfCrime = pd.read_csv('/Users/tanisha/Desktop/cpd-incidents.csv .csv')
print(dfCrime.shape)
dfCrime.head()
```

```python
list(dfCrime.columns)
```

```python
dfCrime['date_from'] = pd.to_datetime(dfCrime['date_from'])
dfCrime.head()
```

```python
dfCrime = dfCrime.loc[(dfCrime['date_from'] >= '2021-01-01') & (dfCrime['date_from'] < '2022-01-01')]
print(dfCrime.shape)
dfCrime.head()
```

```python
list(dfCrime['crime_type'].unique())
```

```python
dfCrime = dfCrime[['lat', 'lon']]

# make our lon and lat into a point type for spacial join
dfCrime['geometry'] = dfCrime.apply(lambda row: Point(row['lon'], row['lat']), axis=1)
dfCrime = dfCrime[['geometry']]
dfCrime.head()
```
