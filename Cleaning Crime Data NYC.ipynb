```python
import pandas as pd
import numpy as np
import geopandas as gpd
from shapely.geometry import Point
```

```python
dfCrime = pd.read_csv('/Users/austincoffelt/Downloads/NYPD_Arrests_Data__Historic__20250412.csv')
print(dfCrime.shape)
dfCrime.head()
```

```python
violentCrimes = ['FELONY ASSAULT', 'RAPE', 'MURDER & NON-NEGL. MANSLAUGHTE', 'FELONY SEX CRIMES', 'ROBBERY']
dfCrime = dfCrime[dfCrime['OFNS_DESC'].isin(violentCrimes)]
print(dfCrime.shape)
dfCrime.head()
```

```python
dfCrime = dfCrime[['PD_DESC', 'Longitude', 'Latitude']]
dfCrime.head()
```

```python
# make our lon and lat into a point type for spacial join
dfCrime['geometry'] = dfCrime.apply(lambda row: Point(row['Longitude'], row['Latitude']), axis=1)
dfCrime = dfCrime[['PD_DESC', 'geometry']]
print(dfCrime.shape)
dfCrime.head()
```

```python
# Read shapefile (automatically loads .shp, .shx, .dbf)
gdf = gpd.read_file("cb_2022_us_bg_500k/cb_2022_us_bg_500k.shp")

# View the first 5 rows
print(gdf.shape)
gdf.head()
```

```python
# make sure points match the crs of gdf
gdfPoints = gpd.GeoDataFrame(
    dfCrime, 
    geometry='geometry',
    crs="EPSG:4269"  # WGS84 coordinate reference system
)
gdfPoints.head()
```

```python
# Basic spatial join (points within polygons)
result = gpd.sjoin(
    gdfPoints, 
    gdf,
    how="left",       # 'left', 'right', or 'inner'
    predicate="within" # Spatial relationship
)
print(result.shape)
result.head()
```

```python
crime = result['AFFGEOID'].value_counts()
crime
```

```python
# export as a csv
crime.to_csv('NY_Violent_Crime_BG.csv')
```
