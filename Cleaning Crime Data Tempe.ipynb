```python
import pandas as pd
import numpy as np
import geopandas as gpd
from shapely.geometry import Point
```

```python
dfCrime = pd.read_csv('/Users/austincoffelt/Downloads/Rent-Walk-Local/General_Offenses_(Open_Data)_7435779472384193066.csv')
print(dfCrime.shape)
dfCrime.head()
```

```python
dfCrime = dfCrime[dfCrime['OccurrenceYear'] == 2018]
print(dfCrime.shape)
dfCrime.head()
```

```python
list(dfCrime.columns)
```

```python
sorted(list(dfCrime['OffenseCustom'].unique()))
```

```python
violentCrimes = ['AGGRAVATED ASSAULT                                                              ',
                 'AGGRAVATED ASSAULT DV                                                           ',
                 'AGGRAVATED ASSAULT-POLICE OFC                                                   ',
                 'HOMICIDE                                                                        ',
                 'MANSLAUGHTER                                                                    ',
                 'MOLESTATION OF A CHILD                                                          ',
                 'ROBBERY-ARMED                                                                   ',
                 'ROBBERY-STRONG ARM                                                              ',
                 'SEXUAL ABUSE                                                                    ',
                 'SEXUAL ABUSE/FONDLING                                                           ',
                 'SEXUAL ASSAULT                                                                  ']
dfCrime = dfCrime[dfCrime['OffenseCustom'].isin(violentCrimes)]
print(dfCrime.shape)
dfCrime.head()
```

```python
dfCrime = dfCrime[['OffenseCustom', 'Latitude', 'Longitude']]

# make our lon and lat into a point type for spacial join
dfCrime['geometry'] = dfCrime.apply(lambda row: Point(row['Longitude'], row['Latitude']), axis=1)
dfCrime = dfCrime[['OffenseCustom', 'geometry']]
dfCrime.head()
```

```python
# Read shapefile (automatically loads .shp, .shx, .dbf)
gdf = gpd.read_file("/Users/austincoffelt/Downloads/Rent-Walk-Local/cb_2018_04_bg_500k/cb_2018_04_bg_500k.shp")

# View the first 5 rows
print(gdf.shape)
gdf.head()
```

```python
# make sure points match the crs of gdf
gdfPoints = gpd.GeoDataFrame(
    dfCrime, 
    geometry='geometry',
    crs="EPSG:4269"  # WGS84 coordinate reference system
)
gdfPoints.head()
```

```python
# Basic spatial join (points within polygons)
result = gpd.sjoin(
    gdfPoints, 
    gdf,
    how="left",       # 'left', 'right', or 'inner'
    predicate="within" # Spatial relationship
)
print(result.shape)
result.head()
```

```python
crime = result['AFFGEOID'].value_counts()
crime
```

```python
# export as a csv
crime.to_csv('Tempe_Violent_Crime_BG.csv')
```
